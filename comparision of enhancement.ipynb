```python
import cv2
import numpy as np
import os
```

```python
video_path = "20s.mp4"
cap = cv2.VideoCapture(video_path)
```

```python
fps = cap.get(cv2.CAP_PROP_FPS)
frame_count = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
duration = frame_count / fps

print("Video Information:")
print(f"Resolution: {width} x {height}")
print(f"FPS: {fps}")
print(f"Total Frames: {frame_count}")
print(f"Duration (seconds): {duration:.2f}")
```

```python
sample_interval = int(fps)  # 1 frame per second
brightness_list = []
contrast_list = []
blur_list = []

frame_id = 0
```

```python
while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break

    if frame_id % sample_interval == 0:
        gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

        # Brightness (mean intensity)
        brightness = np.mean(gray)
        brightness_list.append(brightness)

        # Contrast (standard deviation)
        contrast = np.std(gray)
        contrast_list.append(contrast)

        # Blur estimation (Laplacian variance)
        blur = cv2.Laplacian(gray, cv2.CV_64F).var()
        blur_list.append(blur)

    frame_id += 1
```

```python
cap.release()
```

```python
print("\nDataset Quality Analysis:")

print(f"Average Brightness: {np.mean(brightness_list):.2f}")
print(f"Brightness Variance: {np.var(brightness_list):.2f}")

print(f"Average Contrast: {np.mean(contrast_list):.2f}")

print(f"Average Sharpness (Blur Score): {np.mean(blur_list):.2f}")
```
